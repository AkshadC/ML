## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
data = pd.read_csv('Countries-exercise.csv')
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
seaborn.scatterplot(data['Longitude'],data['Latitude'])
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
X = data.iloc[:,1:3]
```

## Clustering

Assume there are only two clusters. 

```python
Kmeans = KMeans(7)
```

```python
Kmeans.fit(X)
```

### Clustering Resutls

```python
cluster_result = Kmeans.fit_predict(X)
```

```python
cluster_result
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
df_with_clusters = data.copy()
df_with_clusters['Clusters'] = cluster_result
df_with_clusters
```

```python
plt.scatter(df_with_clusters['Longitude'],df_with_clusters['Latitude'],c=df_with_clusters['Clusters'],cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
