Plot the Normal Distribution for class test result of a particular subject. Identify the
Skewness and Kurtosis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
import warnings
warnings.filterwarnings('ignore')
seaborn.set()
```

```python
data1 = pd.read_csv('StudentsPerformance.csv')
data1.describe()
```

```python
data1.head()
```

```python
minimum = data1['Score'].min()
maximum = data1['Score'].max()
ran = maximum - minimum
mean = data1['Score'].mean()
median = data1['Score'].median()
st_dev = data1['Score'].std()
skew = data1['Score'].skew()
kurt = data1['Score'].kurtosis()

points = mean - st_dev, mean + st_dev


seaborn.distplot(data1['Score'], hist=True, kde=True)
seaborn.lineplot(points,[0,0], lw=6,color='yellow', label='std_dev')
plt.title(f'std_dev = {round(st_dev,2)} , mean = {mean} , skewness = {round(skew,2)} , kurtosis = {round(kurt,2)}' )
```
