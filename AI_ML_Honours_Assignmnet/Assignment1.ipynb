Deviation, Covariance, Correlation and Standard error with respect to any example.
Display graphically the distribution of samples.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
import warnings
warnings.filterwarnings("ignore")
seaborn.set()
```

```python
data1 = [9,27,5,49,39,53,48, 42, 6, 57, 50, 13,15, 1, 2, 30, 12, 27,39, 50, 9, 22, 23, 43,22, 49, 1, 21, 20, 4]
data1 = np.array(data1)
data2=  [35 ,39 ,3 ,36 ,45 ,17,40 ,19 ,4 ,49 ,5 ,54,13 ,38 ,37 ,21 ,36 ,52,12 ,9, 7 ,26 ,54 ,41,53 ,4 ,48 ,38 ,58 ,30]
data2= np.array(data2)
```

$Mean = Sum/TotalObservations(n)$

```python
def calc_mean(dataset):
    return dataset.sum()/len(dataset)
```

$Variance = Summation ((x_i - mean)^2 )/n-1 $

```python
def calc_variance(dataset,mean):
    squared_diff = np.square(dataset-mean)
    return calc_mean(squared_diff)
```

$Std Dev = sqrt(variance)$

```python
def calc_standard_dev(dataset,mean):
    variance = calc_variance(dataset,mean)
    return np.sqrt(variance)
```

$co-variance = (Summation ( x_i - mean_x)*(y_i - mean_y) ) / n$

```python
def calc_covariance(data1,data2):
    mean1 = calc_mean(data1)
    mean2 = calc_mean(data2)
    return np.sum(np.multiply(data1 - mean1,data2 - mean2))/len(data1)
```

$Correlation = Co-variance(x,y)/(std_dev(x) * std_dev(y))$

```python
def calc_correlation(data1,data2):
    mean1 = calc_mean(data1)
    mean2 = calc_mean(data2)
    num = np.sum (np.multiply(data1-mean1, data2-mean2))
    de = np.multiply(np.sum (np.square (data1-mean1)), np.sum (np.square (data2-mean2)))
    return num/np.sqrt(de)
```

$Standard Error = standard deviation / sqrt(n)$

```python
def calc_std_error(dataset):
    sd = calc_standard_dev(dataset,calc_mean(dataset))
    return sd/np.sqrt(len(dataset))
```

```python
mean1 = calc_mean(data1)
mean2 = calc_mean(data2)

variance1 = calc_variance(data1,mean1)
variance2 = calc_variance(data1,mean2)

standard_deviation = calc_standard_dev(data1,mean1)

covariance = calc_covariance(data1,data2)

correlation = calc_correlation(data1,data2)

std_error = calc_std_error(data1)

print(" Mean of dataset 1 : ", mean1,"\n",
      "Mean of dataset 2 : ", mean2,"\n",
      "Variance of dataset 1 : ", variance1,"\n",
      "Variance of dataset 2 : ", variance2,"\n",
      "Standard Deviation of both the datasets : ", standard_deviation,"\n"
      " Covariance of both the datasets : ",covariance,"\n"
      " Correlation of both the datasets : ",correlation,"\n",
      "Standard Error : ",std_error,"\n")
```
